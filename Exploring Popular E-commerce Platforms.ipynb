```python
# This will show list of rows which have null value
# in this case Amount is containg 12 null values, status and unnames
pd.isnull(data).sum()
```

```python
# filter out the null value present in the data set

rows_with_nulls=data[data.isnull().any(axis=1)]
rows_with_nulls
```

```python
# we have visualise the data ,so to move further we need to assign the data to the missing columns
```

```python
# i have taken the avg of (Myntra_Product) and based on it assigned the value to the missing data
data['Myntra_Product']=data['Myntra_Product'].fillna(data['Myntra_Product'].mean())
data
```

```python
data['Amazon_Product']=data['Amazon_Product'].fillna(data['Amazon_Product'].mean())
data
```

```python
# all the values are non null as we have assigned
data.info()
```

```python
# these are the city present in the data
grouped_by_city = data.groupby('City')
```

```python
# count of people from a city
people_count_per_city = data.groupby('City').size().reset_index(name='count')
print("Number of people from each city:")
print(people_count_per_city)
```

```python
# Group by city and count the number of people in each city
people_count_per_city = data.groupby('City').size().reset_index(name='count')

# Create the plot
plt.figure(figsize=(14,8))
plt.bar(people_count_per_city['City'], people_count_per_city['count'], color='skyblue')

# Add titles and labels
plt.title('Number of People from Each City')
plt.xlabel('City')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
data = pd.DataFrame(data)
print("Original DataFrame:")
data
```

```python
flipkart_count = data['Flipkart_Product'].sum()
myntra_count = data['Myntra_Product'].sum()
amazon_count = data['Amazon_Product'].sum()

platform_counts = pd.DataFrame({
    'Platform': ['Flipkart', 'Myntra', 'Amazon'],
    'Count': [flipkart_count, myntra_count, amazon_count]
})
print("\nPlatform usage count:")
print(platform_counts)
```

```python
# Create the plot with an increased size
plt.figure(figsize=(12, 6))
plt.bar(platform_counts['Platform'], platform_counts['Count'], color=['cyan', 'cornflowerblue', 'palegreen'])

# Add titles and labels
plt.title('Number of Products Used on Each Platform', fontsize=16)
plt.xlabel('Platform', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Customize the tick labels
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```
